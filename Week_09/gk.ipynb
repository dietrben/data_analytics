```python
import pandas as pd
import random

# Number of pets
num_pets = 4000

# Defining the pet types
pet_types = ['dog', 'cat', 'fish', 'bird']

# Generating random data for each pet
data = {
    'type': [random.choice(pet_types) for _ in range(num_pets)],
    'age': [random.randint(0, 20) for _ in range(num_pets)],  # Assuming pet age in years
    'color': [random.choice(['black', 'white', 'brown', 'grey', 'spotted', 'striped']) for _ in range(num_pets)],
    'weight': [round(random.uniform(0.1, 100), 2) for _ in range(num_pets)],  # Weight in kilograms
    'height': [round(random.uniform(5, 150), 2) for _ in range(num_pets)],  # Height in centimeters
    'eats_meat': [random.choice(['yes', 'no']) for _ in range(num_pets)]
}

# Creating the DataFrame
pets_df = pd.DataFrame(data)
pets_df.head()
```
