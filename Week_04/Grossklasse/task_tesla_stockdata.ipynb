```python
import pandas as pd
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Get the data
today = datetime.now().strftime("%Y-%m-%d")
df = yf.download('TSLA', start = "2022-01-01", end = today, progress=False)
df
```

```python
df['Close'].describe()
```

```python
p10 = df['Close'].quantile(0.10)
p90 = df['Close'].quantile(0.90)

print(f"The 10th percentile is: {p10}")
print(f"The 90th percentile is: {p90}")
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Close Price')
plt.title('Tesla Close Prices with Percentiles')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
